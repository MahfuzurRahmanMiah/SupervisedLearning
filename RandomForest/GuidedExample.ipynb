```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Read in csv from dropbox
y2015 = pd.read_csv(
    'https://www.dropbox.com/s/0so14yudedjmm5m/LoanStats3d.csv?dl=1',
    skipinitialspace=True,
    header=1
)

# Note the warning about dtypes.
```

```python
# let's see what's inside
y2015.head()
```

```python
# how many unique values are there in columns with "object" types?
categorical = y2015.select_dtypes(include=['object'])
for i in categorical:
    column = categorical[i]
    print(i)
    print(column.nunique())
```

```python
# let's get rid of columns with too many unique variables 
# and also change columns that could be numeric into numeric types.


# Convert ID and Interest Rate to numeric.
y2015['id'] = pd.to_numeric(y2015['id'], errors='coerce')
y2015['int_rate'] = pd.to_numeric(y2015['int_rate'].str.strip('%'), errors='coerce')

# Drop other columns with many unique variables
y2015.drop(['url', 'emp_title', 'zip_code', 'earliest_cr_line', 'revol_util',
            'sub_grade', 'addr_state', 'desc'], 1, inplace=True)
```

```python
# what caused an issue in the ID column? NaNs?
y2015.tail()
```

```python
# Remove two summary rows at the end that don't actually contain data.
y2015 = y2015[:-2]
```

```python
# now make dummies for the remaining categorical columns
pd.get_dummies(y2015)
```

```python
# make a RandomForestClassifier now that the data is properly prepared

from sklearn import ensemble
from sklearn.model_selection import cross_val_score

rfc = ensemble.RandomForestClassifier()
X = y2015.drop('loan_status', 1)
Y = y2015['loan_status']
X = pd.get_dummies(X)
X = X.dropna(axis=1)

cross_val_score(rfc, X, Y, cv=10)
```

You'll want to do a few things in this process. First, dive into the data that we have and see which features are most important. This can be the raw features or the generated dummies. You may want to use PCA or correlation matrices.

Can you do it without using anything related to payment amount or outstanding principal? How do you know?

```python
# transform categorical target to numerical for correlation and 
# choose to only accept absolute value of correlation > .2
from sklearn.preprocessing import LabelEncoder 
LE = LabelEncoder()
y2015['loan_status_t'] = LE.fit_transform(y2015['loan_status'])
corr = y2015.corr()
corr_list = corr['loan_status_t'][np.abs(corr['loan_status_t']) > 0.2]
corr_list
```

```python
import seaborn as sns
corr2 = y2015[corr_list.index].corr()
sns.heatmap(corr2, xticklabels=corr2.columns, yticklabels=corr2.columns, cmap='coolwarm')
```

We see that the only categories that satisfy this low bar for correlation are related to payment, outstanding principle, total_rec_prcp and last payment amount. This method of checking correlations informs us that these columns are crucial to any model we make so we can't really model without columns relating to payment amount or outstanding principal.
