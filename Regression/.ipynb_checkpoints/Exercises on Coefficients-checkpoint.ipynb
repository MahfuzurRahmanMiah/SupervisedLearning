### 1. Interpretation and significance

Suppose that we would like to know how much families in the US are spending on recreation annually. We've estimated the following model:

$$ expenditure = 873 + 0.0012annual\_income + 0.00002annual\_income^2 - 223.57have\_kids $$

*expenditure* is the annual spending on recreation in US dollars, *annual_income* is the annual income in US dollars, and *have_kids* is a dummy variable indicating the families with children. Interpret the estimated coefficients. What additional statistics should be given in order to make sure that your interpretations make sense statistically. Write up your answer.

We need a summary showing the p values associated with these coefficients. That will tell us how useful these coefficients in terms of contribution to to the target variable. 

If all the coefficients are statistically significant, we can state the following:
873 is the constant value and is also know as the bias. Technically, it assumes a base level of $873 dollars of recreactional expenditure even when a family has no money or kids. More could be stated if we knew more about the variables being used to develop this model.

Families spend, on average, $223.57 dollars less than families without kids on recreational expenditure. 

In relation to annual income, we see that the variable appears twice. Because this variable is squared, we realize that annual income has a quadratic relationship to recreational expenditure. The graph below shows how annual income relates to recreational expenditure, separately considering families with and without children.

```python
import numpy as np
import matplotlib.pyplot as plt

# X represents various annual incomes in $1000 steps 
X = np.arange(20000, 100000, 1000)
# Y represents annual expenditure on recreation
# for families with no children
Y = 873 + 0.0012*X + 0.000002*(X**2)
# Y_kids represents annual expenditure on recreation
# for families with children
Y_kids = (873 - 223.57) + 0.0012*X + 0.000002*(X**2)


plt1 = plt.plot(X, Y, label="with no children")
plt2 = plt.plot(X, Y_kids, label="with children")
plt.xlabel("income")
plt.ylabel("recreation expenditure")
plt.title("The relation between income and recreation expenditure")
plt.legend()
plt.show()
```

From this graph actually, one might even be able to extrapolate that the "with or without children" factor doesn't really matter when we consider that annual family incomes usually are more than $20,000. The lines are practically overlapping.

### 2. Weather model

* First, load the dataset from the **weatherinszeged** table from Thinkful's database.

Now, estimate your model using OLS. Are the estimated coefficients statistically significant? Are the signs of the estimated coefficients in line with your previous expectations?
```python
# import necessary libraries, set up postgres credentials
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sqlalchemy import create_engine
import statsmodels.api as sm

import warnings
warnings.filterwarnings('ignore')

postgres_user = 'dsbc_student'
postgres_pw = '7*.8G9QH21'
postgres_host = '142.93.121.174'
postgres_port = '5432'
postgres_db = 'weatherinszeged'
```

```python
# run the engine to extract your dataframe, close the connection
engine = create_engine('postgresql://{}:{}@{}:{}/{}'.format(
    postgres_user, postgres_pw, postgres_host, postgres_port, postgres_db))
weather_df = pd.read_sql_query('select * from weatherinszeged',con=engine)

# no need for an open connection, as we're only doing a single query
engine.dispose()
```

```python
# Y is the target variable
Y = weather_df['apparenttemperature'] - weather_df['temperature']
# X is the feature set
X = weather_df[['humidity','windspeed']]

X = sm.add_constant(X)

results = sm.OLS(Y, X).fit()

results.summary()
```

- p-values of all coefficients are less than .05
- From the constant, we gather than at baseline, without taking into consideration humidity and windspeed, the temperature is always higher than the apparent temperature by 2.438 degrees.
- One may think that increased humidity would increase the difference between the 'apparenttemperature' and 'temperature'. The negative sign on our model shows that increased humidity would reflect as lower apparent temperatures in relation to the temperature. The sign for windspeed is also negative which also means that increases in windspeed reflect as lower apparent temperatures in relation to temperature, which is to be expected.
- Our model suggests a one point deviation in humidity sways temperature more than a one point deviation in windspeed.

```python
weather_df['humidity_windspeed_interaction'] = weather_df.humidity * weather_df.windspeed

# Y is the target variable
Y = weather_df['apparenttemperature'] - weather_df['temperature']
# X is the feature set
X = weather_df[['humidity','windspeed', 'humidity_windspeed_interaction']]

X = sm.add_constant(X)

results = sm.OLS(Y, X).fit()

results.summary()
```

- Like the previous model, all the coefficients still have p-values less than .05.
- Including the interaction changed the sign for the humidity and windspeed coefficients.
- The interaction between humidity and windspeed weighs more on the model, per unit of deviation, in relation to one-point deviations in humidity and windpseed alone. 
- A one point change in humidity effectively results in 0.178 - (0.297 * windspeed) increase in the target. When windspeed is high, humidity effects are decreased.
- A one point change in windspeed effectively results in 0.091 - (0.297 * humidity) increase in the target. When humidity is high, any windspeed contribution is effectively decreased. So the two variables mitigate the effects of each other.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sqlalchemy import create_engine
import statsmodels.api as sm

import warnings
warnings.filterwarnings('ignore')

postgres_user = 'dsbc_student'
postgres_pw = '7*.8G9QH21'
postgres_host = '142.93.121.174'
postgres_port = '5432'
postgres_db = 'houseprices'
```

```python
engine = create_engine('postgresql://{}:{}@{}:{}/{}'.format(
    postgres_user, postgres_pw, postgres_host, postgres_port, postgres_db))
house_prices_df = pd.read_sql_query('select * from houseprices',con=engine)

# no need for an open connection, as we're only doing a single query
engine.dispose()
```

```python
house_prices_df = pd.concat([house_prices_df,pd.get_dummies(house_prices_df.mszoning, prefix="mszoning", drop_first=True)], axis=1)
house_prices_df = pd.concat([house_prices_df,pd.get_dummies(house_prices_df.landcontour, prefix="landcontour", drop_first=True)], axis=1)
dummy_column_names = list(pd.get_dummies(house_prices_df.mszoning, prefix="mszoning", drop_first=True).columns)
dummy_column_names = dummy_column_names + list(pd.get_dummies(house_prices_df.landcontour, prefix="landcontour", drop_first=True).columns)
```

```python
# Y is the target variable
Y = house_prices_df['saleprice']
# X is the feature set
X = house_prices_df[['overallqual','grlivarea','garagecars','garagearea','totalbsmtsf','firstflrsf'] + dummy_column_names]


X = sm.add_constant(X)

results = sm.OLS(Y, X).fit()

results.summary()
```


The statistically significant variables are overallqual, grlivarea, garagecars, totalbsmtsf, mszoning_FV and mszoning_RL, landcontour_HLS,landcontour_Low,landcontour_Lvl. All the other variables are statistically insignificant. According to the estimation results:

1 point increase in overallqual results in \$2331 increase in sale price.
1 point increase in grlivarea results in \$45.60 increase in sale price.
1 point increase in garagecars results in \$12680 increase in sale price.
1 point increase in totalbsmtsf results in \$22.78 increase in sale price.
Average sale price is higher in mszoning_FV and mszoning_RL than the base zone by \$15470 and \$16080 respectively. So, the average house price is the highest in mszoning_RL.
Average sale price is higher if landcontour is \_HLS, \_Low, and \_Lvl instead of \_Bnk by \$37950, \$42510, and \$19120 respectively. 
Now let's exclude statistically insignificant variables from the model and re-estimate it:

```python
# Y is the target variable
Y = house_prices_df['saleprice']
# X is the feature set
X = house_prices_df[['overallqual', 'grlivarea', 'garagecars', 'totalbsmtsf', 'mszoning_FV', 'mszoning_RL', 'landcontour_HLS','landcontour_Low','landcontour_Lvl']]

X = sm.add_constant(X)

results = sm.OLS(Y, X).fit()

results.summary()
```

The results resemble the previous model's results. The coefficients of the dummies changed relatively larger than the other variables.

**Considering the results, the most prominent factors affecting the sale price seem to be the landcontour, overallqual and the mszoning.**
